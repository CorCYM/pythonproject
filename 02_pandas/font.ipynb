```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.random.randint(-100,100,50).cumsum()

plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```

```python
##1. 폰트파일로 FontProperties()
import matplotlib.font_manager as fm

path = 'D2CODING.TTF'
fontprop = fm.FontProperties(fname=path, size=14)

data = np.random.randint(-100,100,50).cumsum()
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이', fontproperties = fontprop)
plt.ylabel('주식가격', fontproperties = fontprop)
plt.xlabel('시간(분)', fontproperties = fontprop)
```

```python
#2.rcParams[]를 설정

import matplotlib

print('설정파일위치:',matplotlib.matplotlib_fname())
print('기본 설정 폰트:',plt.rcParams['font.family'])
print('기본 폰트 크기:',plt.rcParams['font.size'])
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100,100,50).cumsum()
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```
