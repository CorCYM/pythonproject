```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/data/gapminder.tsv', sep='\t')
```

```python
df.groupby('year').lifeExp.mean()
```

```python
df.groupby('year')['lifeExp'].mean()
```

```python
df.groupby('year')['lifeExp'].mean().to_frame()
```

```python
df.groupby('year')['lifeExp'].mean().to_frame().reset_index()
```

```python
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum/n
```

```python
df.groupby('year')['lifeExp'].agg(my_mean) ## 사용자가 직접 함수를 만들어 적용가능
```

```python
df.groupby('year')['lifeExp'].agg([np.count_nonzero, np.mean, np.std])
```

```python
df.groupby('year').agg({'lifeExp':'mean',
                        'pop':'median', 
                        'gdpPercap':'median'}) 
```

```python
df.groupby('year').agg({'lifeExp':'mean',
                        'lifeExp':'median',
                        'pop':'median', 
                        'gdpPercap':'median'}) ## 키 값이 같을 경우 뒤에 선언된 것으로 덮어 씀
```
